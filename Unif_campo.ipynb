## Preparar los datos para machine learning

### Subtask:
Asegurarse de que la columna de tiempo esté en un formato adecuado para series temporales, identificar los intervalos de tiempo donde faltan datos en el DataFrame filtrado y crear una estructura de datos que combine los tiempos existentes con los tiempos donde se realizará la imputación.


**Reasoning**:
Verify the time column data type, identify time gaps, create a complete time index, and reindex the filtered DataFrame to prepare it for imputation.



```python
# 1. Verify the data type of the 'time' column
print(df_filtered_ml['time'].dtype)

# 2. Identify the gaps in the 'time' column
time_diff = df_filtered_ml['time'].diff()
print("Time differences between consecutive data points:")
print(time_diff.describe())
# Identify large gaps (e.g., significantly larger than the median/mean difference)
large_gaps = time_diff[time_diff > time_diff.median() * 2] # Using median*2 as a simple heuristic
print("\nIndices and values of large time gaps:")
print(large_gaps)

# 3. Create a complete time series index
# Assuming the data has a relatively consistent frequency, we can infer it from the typical difference
# Using the median difference to be less sensitive to outliers
typical_interval = time_diff.median()
start_time = df_filtered_ml['time'].min()
end_time = df_filtered_ml['time'].max()

# Create a date_range based on the start, end, and inferred frequency
# Convert float time to datetime objects, assuming time is in seconds from a reference point (e.g., start_time)
# A simple approach is to use the row index as a proxy for time steps if intervals are irregular but data is ordered
# However, the 'time' column is already float, which is suitable for numerical operations.
# Let's create a regular numerical sequence based on the start, end, and typical interval.
# We'll need to merge or reindex based on this numerical time.

# A more robust way to create a complete time index for imputation is to use pandas' date_range
# We need to convert the float time to datetime for date_range
# Let's assume the 'time' column represents seconds from the start of the recording.
start_datetime = pd.to_datetime('2025-08-11 18:08:13') + pd.to_timedelta(df_filtered_ml['time'].min(), unit='s')
end_datetime = pd.to_datetime('2025-08-11 18:08:13') + pd.to_timedelta(df_filtered_ml['time'].max(), unit='s')

# Determine the frequency based on the most frequent difference or median difference
# Let's use the median difference as a frequency approximation
freq = pd.to_timedelta(typical_interval, unit='s')

complete_time_index_datetime = pd.date_range(start=start_datetime, end=end_datetime, freq=freq)

# Convert the original 'time' column to datetime for merging/reindexing
df_filtered_ml['datetime'] = pd.to_datetime('2025-08-11 18:08:13') + pd.to_timedelta(df_filtered_ml['time'], unit='s')

# 4. Reindex df_filtered_ml using this complete time index
df_reindexed = df_filtered_ml.set_index('datetime').reindex(complete_time_index_datetime)

# Display the head of the reindexed DataFrame to show introduced NaNs
display(df_reindexed.head())
```

Analizar los datos de campo magnético de un archivo CSV, convertir los valores de microteslas a militeslas, generar un reporte, visualizar la relación entre campo magnético y tiempo con diferentes granularidades de tiempo (1s, 5s, 10s), filtrar los datos para incluir solo valores mayores a 1.9 mT, y utilizar machine learning para predecir los valores faltantes en el rango de tiempo de 25 a 30 segundos basándose en los datos hasta los 25 segundos, y finalmente visualizar los datos originales y predichos. El archivo CSV se llama "medicion_campo - medicion_campo.csv".

## Preparar los datos para pronóstico

### Subtask:
Crear un nuevo DataFrame que contenga solo las filas donde el campo magnético total (BT) es mayor a 1.9 mT, dividir los datos filtrados en un conjunto de entrenamiento (0 a ~25 segundos) y un conjunto de validación/prueba (después de ~30 segundos), y asegurar que la columna de tiempo esté en un formato adecuado para modelos de series temporales.


**Reasoning**:
Filter the DataFrame, split it into training and test sets based on time, and verify the time column data type.



```python
# 1. Filter the DataFrame for BT > 1.9 mT
df_filtered_forecast = df_magnetic_field[df_magnetic_field['BT'] > 1.9].copy()

# 2. Divide the filtered data into training and test sets
df_train_forecast = df_filtered_forecast[df_filtered_forecast['time'] <= 25].copy()
df_test_forecast = df_filtered_forecast[df_filtered_forecast['time'] >= 30].copy()

# 3. Verify the data type of the 'time' column
print("Data type of 'time' in df_train_forecast:", df_train_forecast['time'].dtype)
print("Data type of 'time' in df_test_forecast:", df_test_forecast['time'].dtype)

# 4. Display the head of the train and test DataFrames
print("\nHead of df_train_forecast:")
display(df_train_forecast.head())

print("\nHead of df_test_forecast:")
display(df_test_forecast.head())
```

## Seleccionar y entrenar un modelo de pronóstico

### Subtask:
Elegir un modelo apropiado para pronóstico de series temporales (por ejemplo, ARIMA, Prophet, modelos basados en Machine Learning como Random Forest o Gradient Boosting con características de tiempo), y entrenar el modelo con el conjunto de datos de entrenamiento (`df_train_forecast`).


**Reasoning**:
Train a RandomForestRegressor model using the 'time' column as the feature and the 'BT' column as the target from the training dataset.



```python
from sklearn.ensemble import RandomForestRegressor

# Prepare the training data
X_train = df_train_forecast[['time']]
y_train = df_train_forecast['BT']

# Choose and train the model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# The model is now trained and stored in the 'model' variable.
```

## Realizar predicciones

### Subtask:
Utilizar el modelo entrenado para predecir los valores del campo magnético en los intervalos de tiempo donde faltan datos.


**Reasoning**:
To identify the missing time intervals between 25 and 30 seconds, I will create a complete time sequence for this range based on the typical interval found earlier, and then identify the times within this sequence that are not present in the original filtered data (`df_filtered_forecast`). Finally, I will prepare these missing time points for prediction using the trained model.



```python
import numpy as np

# Identify the typical time interval from the original data if not already done
# (Assuming typical_interval was calculated in a previous step and is available)
# If not, calculate it:
if 'typical_interval' not in globals():
    time_diff_all = df_magnetic_field['time'].diff()
    typical_interval = time_diff_all.median()

# 1. Create a complete time sequence between 25 and 30 seconds
start_forecast_time = 25.0
end_forecast_time = 30.0

# Generate a sequence of time points with the typical interval
# Use arange with a small tolerance for floating point comparisons
complete_forecast_times = np.arange(start_forecast_time, end_forecast_time + typical_interval, typical_interval)

# 2. Identify times where data is missing in the filtered DataFrame within this range
# Get the existing times in the filtered DataFrame within the forecast range
existing_times_filtered = df_filtered_forecast[(df_filtered_forecast['time'] >= start_forecast_time) &
                                               (df_filtered_forecast['time'] <= end_forecast_time)]['time'].values

# Find the times in the complete sequence that are not in the existing filtered times
# Use isin with a tolerance for floating point comparison
missing_forecast_times = complete_forecast_times[~np.isclose(complete_forecast_times[:, None], existing_times_filtered, atol=typical_interval/2).any(axis=1)]

# 3. Create a DataFrame with these missing time intervals for prediction
df_predict_times = pd.DataFrame({'time': missing_forecast_times})

# 4. Use the trained model to predict the magnetic field (BT) values for these times
# The model was trained on X_train which was df_train_forecast[['time']]
X_predict = df_predict_times[['time']]
predicted_bt = model.predict(X_predict)

# Store the predictions in the DataFrame
df_predict_times['BT_predicted'] = predicted_bt

# Display the first few predicted values
display(df_predict_times.head())
```
